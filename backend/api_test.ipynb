```python
word_list = ["listen", "silent", "enlist", "hello", "world", "dog", "god"]

def filter_len_wordlist(word, wordlist):
    filtered_wordlist = []
    for comp_word in wordlist:
        if len(word) == len(comp_word):
            filtered_wordlist.append(comp_word)
    return filtered_wordlist



filter_len_wordlist("lol", word_list)

def find_anagram_pairs(wordlist):

    anagram_tuple_list = []

    for word in wordlist:
        filtered_len_wordlist = filter_len_wordlist(word, wordlist)
        for filtered_word in filtered_len_wordlist:
            matching_letter_count = 0
            for letter in word:
                if letter in filtered_word:
                    matching_letter_count += 1
            if matching_letter_count == len(word):
                anagram_tuple_list.append((word, filtered_word))


    return anagram_tuple_list



 
find_anagram_pairs(word_list)
```
